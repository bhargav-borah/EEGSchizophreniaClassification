```python
from glob import glob 
import os
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_file_paths = glob('data/*.edf')
```

```python
len(all_file_paths)
```

```python
healthy_file_paths = all_file_paths[:14]
schizo_file_paths = all_file_paths[14:]
```

```python
schizo_file_paths
```

```python
def read_data(file_path):
    
    data = mne.io.read_raw_edf(file_path, preload=True)
    data.set_eeg_reference()
    data.filter(l_freq=0.5, h_freq=45)
    epochs = mne.make_fixed_length_epochs(data, duration=5, overlap=1)
    array = epochs.get_data()
    
    return array
```

```python
sample_data = read_data(healthy_file_paths[0])
sample_data
```

```python
sample_data.shape # no. of epochs, no. of channels, length of signal
```

```python
control_epochs_array = [read_data(i) for i in healthy_file_paths]
patient_epochs_array = [read_data(i) for i in schizo_file_paths]
```

```python
control_epochs_array[0].shape
```

```python
patient_epochs_array[0].shape
```

```python
control_epochs_labels = [len(i) * [0] for i in control_epochs_array]
patient_epochs_labels = [len(i) * [1] for i in patient_epochs_array]
```

```python
data_list = control_epochs_array + patient_epochs_array
labels_list = control_epochs_labels + patient_epochs_labels
```

```python
labels_list
```

```python
group_list = [[i] * len(j) for i, j in enumerate(data_list)]
```

```python
data_array = np.vstack(data_list)
labels_array = np.hstack(labels_list)
group_array = np.hstack(group_list)
print(data_array.shape, labels_array.shape, group_array.shape)
```
