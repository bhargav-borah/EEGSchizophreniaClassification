```python
healthy_file_paths = all_file_paths[:14]
schizo_file_paths = all_file_paths[14:]
```

```python
def read_data(file_path):
    
    data = mne.io.read_raw_edf(file_path, preload=True)
    data.set_eeg_reference()
    data.filter(l_freq=0.5, h_freq=45)
    epochs = mne.make_fixed_length_epochs(data, duration=5, overlap=1)
    array = epochs.get_data()
    
    return array
```

```python
sample_data = read_data(healthy_file_paths[0])
sample_data
```

```python
sample_data.shape # no. of epochs, no. of channels, length of signal
```

```python
control_epochs_array = [read_data(i) for i in healthy_file_paths]
patient_epochs_array = [read_data(i) for i in schizo_file_paths]
```

```python
control_epochs_array[0].shape
```

```python
patient_epochs_array[0].shape
```

```python
control_epochs_labels = [len(i) * [0] for i in control_epochs_array]
patient_epochs_labels = [len(i) * [1] for i in patient_epochs_array]
```

```python
data_list = control_epochs_array + patient_epochs_array
labels_list = control_epochs_labels + patient_epochs_labels
```

```python
labels_list
```

```python
group_list = [[i] * len(j) for i, j in enumerate(data_list)]
```

```python
data_array = np.vstack(data_list)
labels_array = np.hstack(labels_list)
group_array = np.hstack(group_list)
print(data_array.shape, labels_array.shape, group_array.shape)
```

#### Feature Extraction

```python
from scipy import stats

def mean(x):
    return np.mean(x, axis=-1)

def std(x):
    return np.mean(x, axis=-1)

def ptp(x):
    return np.ptp(x, axis=-1)

def var(x):
    return np.var(x, axis=-1)

def minim(x):
    return np.min(x, axis=-1)

def maxim(x):
    return np.max(x, axis=-1)

def argminim(x):
    return np.argmin(x, axis=-1)

def argmaxim(x):
    return np.argmax(x, axis=-1)

def rms(x):
    return np.sqrt(np.mean(x**2, axis=-1))

def abs_diff_signal(x):
    return np.sum(np.abs(np.diff(x, axis=-1)), axis=-1)
    
def skewness(x):
    return stats.skew(x, axis=-1)

def kurtosis(x):
    return stats.kurtosis(x, axis=-1)

def concatenate_features(x):
    return np.concatenate((mean(x), std(x), ptp(x), 
                          var(x), minim(x), maxim(x), 
                          argminim(x), argmaxim(x), rms(x), 
                          abs_diff_signal(x), skewness(x), kurtosis(x)), axis=-1)
```

```python
features = []
for data in data_array:
    features.append(concatenate_features(data))
```

```python
features_array = np.array(features)
features_array.shape
```

#### Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GroupKFold, GridSearchCV
```

```python
model = LogisticRegression()
gkf = GroupKFold(5)
pipe = Pipeline([('scaler', StandardScaler()), ('clf', model)])
param_grid = {'clf__C': [0.1, 0.5, 0.7, 1, 3, 5, 7]}
gscv = GridSearchCV(pipe, param_grid, cv=gkf, n_jobs=-1)
gscv.fit(features_array, labels_array, groups=group_array)
```
